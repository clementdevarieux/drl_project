```python
import numpy as np
import matplotlib.pyplot as plt
import json
import pandas as pd
import seaborn as sns
```

### DDQN

```python
import os
import ast

# Chemin du dossier contenant les fichiers
folder_path = 'results/TicTacToe/DDQN/'

# Parcourir tous les fichiers du dossier
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        
        # Lecture des données
        with open(file_path, 'r') as f:
            data = f.read()
        
        # Convertir le contenu du fichier en dictionnaire
        try:
            results = ast.literal_eval(data)
        except (ValueError, SyntaxError) as e:
            print(f"Erreur de décodage dans le fichier {filename}: {e}")
            continue
        
        # Extraire les informations nécessaires
        mean_score = results.get('mean_score', 'Non disponible')
        score_evolution = results.get('score_evolution', {})
        params = results.get('params', {})
        
        # Afficher les KPI
        print(filename)
        print(f"Mean Score: {mean_score}")
        print("Score Evolution:",score_evolution)
        print("Params:", params)
        print("\n")
```

```python
import ast
import matplotlib.pyplot as plt
import seaborn as sns

# Lecture des données
file_path = 'results/TicTacToe/DDQN/TicTacToe_DDQN_PER_1000_1733177587.541299.txt'
with open(file_path, 'r') as f:
    data = f.read()

# Convertir le contenu du fichier en dictionnaire
results = ast.literal_eval(data)

# Extraire les informations nécessaires
mean_score = results.get('mean_score', 'Non disponible')
mean_steps = results.get('mean_steps', 'Non disponible')
score_evolution = results.get('score_evolution', {})
step_history = results.get('step_history', {})
params = results.get('params', {})

# Création du DataFrame pour Seaborn
data = pd.DataFrame({
    "Simulations": list(score_evolution.keys()),
    "Scores": list(score_evolution.values()),
    "Steps": list(step_history.values())
})

# Visualisations avec Seaborn
sns.set(style="whitegrid")

# Graphique en ligne avec deux échelles
fig, ax1 = plt.subplots(figsize=(12, 6))

# Tracer les scores sur l'axe de gauche
ax1.set_xlabel("Nombre de simulations")
ax1.set_ylabel("Scores", color="lime")
sns.lineplot(data=data, x="Simulations", y="Scores", label="Scores", marker="o", ax=ax1, color="lime")
ax1.tick_params(axis='y', labelcolor="lime")

# Créer un deuxième axe pour les étapes
ax2 = ax1.twinx()
ax2.set_ylabel("Steps", color="red")
sns.lineplot(data=data, x="Simulations", y="Steps", label="Steps", marker="x", linestyle="--", ax=ax2, color="red")
ax2.tick_params(axis='y', labelcolor="red")

# Titre et légende
params_str = ', '.join([f"{k}: {v}" for k, v in params.items()])
params_str = params_str.replace('batch_size', '\nbatch_size')
plt.title(f"Farkle - Double DQN , 1000 games\nScores et Steps\nMean Score: {mean_score}, Mean Steps: {mean_steps}\nParams: {params_str}")
fig.tight_layout()
plt.show()
```
